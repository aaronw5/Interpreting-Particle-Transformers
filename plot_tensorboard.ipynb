```python
# Plot tensorboard validation loss using json data format

import json
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapath = '/mnt/c/Users/timtl/Downloads/Aug25_10-34-07_timlegge-hls4ml-reg-drop-log-qmk6mHLS4ML_kinplus_ParT_loss_eval.json'

df = pd.read_json(datapath)
```
